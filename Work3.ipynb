```python
def invert_dictionary(input_dict):
    inverted_dict = {}
    for key, value in input_dict.items():
        if value in inverted_dict:
            if isinstance(inverted_dict[value], list):
                inverted_dict[value].append(key)
            else:
                inverted_dict[value] = [inverted_dict[value], key]
        else:
            inverted_dict[value] = key

    return inverted_dict

my_dict = {"a": 1, "b": 2, "c": 4, "d": 3}
inverted = invert_dictionary(my_dict)
print(inverted)
```

```python
def merge_dictionaries(dict1, dict2):
    merged_dict = {}
    for key, value in dict1.items():
        merged_dict[key] = value
    for key, value in dict2.items():
        if key in merged_dict:
            merged_dict[key] += value
        else:
            merged_dict[key] = value

    return merged_dict

dict1 = {"a": 1, "b": 2, "c": 3}
dict2 = {"b": 3, "d": 4, "e": 5}
merged = merge_dictionaries(dict1, dict2)
print(merged)
```

```python
def filter_dictionary_by_value(input_dict, wall):
    filtered_dict = {}
    for key, value in input_dict.items():
        if value > wall:
            filtered_dict[key] = value
    return filtered_dict

my_dict = {"a": 1, "b": 5, "c": 2, "d": 7, "e": 3}
threshold_value = 3
filtered = filter_dictionary_by_value(my_dict, threshold_value)
print(filtered)
```

```python
def find_key_with_max_value(input_dict):
    max_key = None
    max_value = float('-inf')
    for key, value in input_dict.items():
        if value > max_value:
            max_value = value
            max_key = key
    return max_key

my_dict = {"a": 1, "b": 5, "c": 2, "d": 7, "e": 3}
max_key = find_key_with_max_value(my_dict)
print(max_key)
```

```python
def find_key_with_min_value(input_dict):
  min_ = float('inf')
  min_key = None
  for key, value in input_dict.items():
    if value < min_:
      min_ = value
      min_key = key
  return min_key

my_dict = {"a": 1, "b": 5, "c": 2, "d": 7, "e": 3}
min_key = find_key_with_min_value(my_dict)
print(min_key)
```

```python
def remove_keys(input_dict, keys_to_remove):
    for key in keys_to_remove:
        if key in input_dict:
            del input_dict[key]
    return input_dict

my_dict = {"a": 1, "b": 5, "c": 2, "d": 7, "e": 3}
keys_to_remove = ["b", "d", "f"]
modified_dict = remove_keys(my_dict, keys_to_remove)
print(modified_dict)
```

```python
def group_keys_by_length(input_dict):
    grouped_dict = {}
    for key in input_dict:
        key_length = len(key)
        if key_length in grouped_dict:
            grouped_dict[key_length].append(key)
        else:
            grouped_dict[key_length] = [key]
    return grouped_dict

my_dict = {"a": 1, "bb": 5, "ccc": 2, "dd": 7, "e": 3, "ffff": 4}
grouped = group_keys_by_length(my_dict)
print(grouped)
```

```python
def combine_values_to_lists(dict1, dict2):
    combined_dict = {}
    for key, value in dict1.items():
        if key in combined_dict:
            combined_dict[key].append(value)
        else:
            combined_dict[key] = [value]
    for key, value in dict2.items():
        if key in combined_dict:
            combined_dict[key].append(value)
        else:
            combined_dict[key] = [value]

    return combined_dict

dict1 = {"a": 1, "b": 2, "c": 3}
dict2 = {"b": 4, "d": 5, "e": 6}
combined = combine_values_to_lists(dict1, dict2)
print(combined)
```

```python
def sort_dictionary_by_keys(input_dict):
    sorted_keys = sorted(input_dict.keys())
    sorted_list = {}
    for key in sorted_keys:
      sorted_list.update({key: input_dict[key]})
    return sorted_list

my_dict = {"c": 3, "a": 10, "b": 2}
sorted_dict = sort_dictionary_by_keys(my_dict)
print(sorted_dict)
```

```python
def sort_dictionary_by_values(input_dict):
    sorted_items = sorted(input_dict.items())
    sorted_dict = {}
    for key, value in sorted_items:
      sorted_dict[key] = value
    return sorted_dict

my_dict = {"c": 3, "a": 1, "b": 2}
sorted_dict = sort_dictionary_by_values(my_dict)
print(sorted_dict)
```

```python
def create_nested_dictionary():
    nested_dict = {}
    alphabet = ['q', 'w', 'e', 'r', 't', 'y']

    for letter in alphabet:
        nested_dict[letter] = {}
        for number in range(1, 11):
            nested_dict[letter][number] = 'ABOBA'

    return nested_dict
my_nested_dict = create_nested_dictionary()

for letter in ['q','w','e']:
    print(f"{letter}: {my_nested_dict[letter]}")
```

```python
def sum_dictionary_values(input_dict):
    total = 0
    for value in input_dict.values():
        total += value
    return total

my_dict = {"a": 1, "b": 2, "c": 3, "d": 4}
sum_of_values = sum_dictionary_values(my_dict)
print(sum_of_values)
```

```python
def product_dictionary_values(input_dict):
    product = 1
    for value in input_dict.values():
        product *= value
    return product

my_dict = {"a": 1, "b": 2, "c": 3, "d": 4}
product_of_values = product_dictionary_values(my_dict)
print(product_of_values)
```
