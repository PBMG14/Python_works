```python
from math import sqrt
a = int(input())
b = int(input())
c = int(input())
if a+b>c and a+c>b and b+c>a:
  p = (a+b+c)/2
  s = sqrt(p*(p-a)*(p-b)*(p-c))
  print(2*s/a)
  print(2*s/b)
  print(2*s/c)
else:
  print("Triangle doesn't exist")
```
