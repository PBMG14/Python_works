```python
class Transaction:
  def __init__(self, name, check):
    self.name = name
    self.check = check
class Payment(Transaction):
  def __init__(self, name, check, sum):
    super().__init__(name, check)
    self.sum = sum
  def paid(self, money):
    self.sum += money
  def WhatLeft(self):
    print(f'In {self.name} left to pay {self.check - self.sum}')

cafe = Payment('Cafe', 1000, 0)
cafe.paid(100)
cafe.WhatLeft()
cafe.paid(900)
cafe.WhatLeft()
```

```python
from abc import ABC, abstractmethod
class CharElement(ABC):
  @abstractmethod
  def draw():
    pass
class BarChart(CharElement):
  def draw(self):
    print("|")
class LineChart(CharElement):
  def draw(self):
    print("-")



a = LineChart()
b = BarChart()
a.draw()
b.draw()
```

```python
from abc import ABC, abstractmethod

class Payment(ABC):
  def __init__(self, date, sum):
    self.sum = sum
    self.date = date
  @abstractmethod
  def processPayment(self, payment):
    self.sum -= payment

class CreditCardPayment(Payment):
  def __init__(self, date, sum, payment, card_number):
    super().__init__(date, sum)
    self.card_number = card_number
    self.processPayment(payment)

  def processPayment(self, payment):
    super().processPayment(payment)
    print(f"Credit card payment successful, transaction date:{self.date}, number:{self.card_number}")

class PayPalPayment(Payment):
  def __init__(self, date, sum, payment, email):
    super().__init__(date, sum)
    self.email = email
    self.processPayment(payment)

  def processPayment(self, payment):
    super().processPayment(payment)
    print(f"PayPal payment successful, transaction date:{self.date}, number:{self.email}")

credit_card_payment = CreditCardPayment("2024-01-01", 100, 20, "12345")
print(credit_card_payment.sum)

paypal_payment = PayPalPayment("2024-01-02", 50, 30, "ABOBA@mail.ru")
print(paypal_payment.sum)
```
