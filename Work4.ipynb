```python
file = open('/input/input1.txt', 'r')
k = int(input('Введите число учитываемых сиволов в строках: '))
str1 = file.read()
file.close()
str1 = str1.split('\n')
print(str1)
str2 = []
str3 = []

for line in str1:
    if len(line) >= k:
        str2.append(line[:k])
        str3.append(line[k-1])
    else:
        str2.append(line)
        str3.append(' ')

file = open('/input/output1.txt', 'w')
for s in str2:
    file.write(s + '\n')
file.close()
file = open('/input/output2.txt', 'w')
for c in str3:
    file.write(c + '\n')
file.close()
```

```python
import random
class Turn:
    def __init__(self, question, var1, var2, var3, var4, var5, ans):
        self.question = question
        self.var = [var1, var2, var3, var4, var5]
        answer = self.var[int(ans)-1]
        random.shuffle(self.var)
        self.ans = 1+self.var.index(answer)
    def display_question(self):
        print(self.question)
        print(f"1. {self.var[0]}")
        print(f"2. {self.var[1]}")
        print(f"3. {self.var[2]}")
        print(f"4. {self.var[3]}")
        print(f"5. {self.var[4]}")
    def check_answer(self, user_answer):
        return user_answer == self.ans
class Questions:
    def __init__(self, questions_data):
        self.list_of_questions = []
        for item in questions_data:
            parts = item.split(';')
            if len(parts) == 7:
                question, var1, var2, var3, var4, var5, ans = parts
                self.list_of_questions.append(Turn(question, var1, var2, var3, var4, var5, ans))

txt = "Что такое Python?;Интерпретируемый язык программирования;Язык разметки;База данных;Операционная система;Фреймворк;1|Кто является создателем языка Python?;Гвидо ван Россум;Джеймс Гослинг;Бьёрн Страуструп;Дональд Кнут;Денис Ричи;1|Какое ключевое слово используется для определения функции в Python?;def;class;if;for;while;1|Что означает аббревиатура API?;Application Programming Interface;Advanced Programming Interface;Automated Programming Interface;Application Process Interface;Automated Process Interface;1|Какой тип данных является неизменяемым в Python?;Tuple;List;Dictionary;Set;String;1|Какой оператор используется для возведения в степень в Python?;**;*;/;+;%;1|Что такое переменная?;Именованная область памяти для хранения данных;Команда для вывода данных на экран;Функция для выполнения математических операций;Оператор для управления потоком выполнения;Структура данных;1|Как называется процесс исправления ошибок в коде?;Отладка;Компиляция;Интерпретация;Рефакторинг;Тестирование;1|Что делает метод split() в Python?;Разбивает строку на список подстрок;Объединяет список строк в одну строку;Удаляет пробелы из начала и конца строки;Заменяет одну подстроку на другую;Находит индекс подстроки;1|Какой модуль используется для работы со случайными числами в Python?;random;math;os;sys;datetime;1|"
file = open('/input/questions.txt', 'w')
file.write(txt)
file.close()
print("Добро пожаловать на викторину!")
print("Выберите число вопросов..")
q_num = int(input())
print(f"Вопросов выбрано: {q_num}\nДа начнется игра!")
file = open('/input/questions.txt', 'r')
str1 = file.read()
questions_data = str1.split('|')
file.close()
questions = Questions(questions_data)
random.shuffle(questions.list_of_questions)
sum = 0
q_num = min(q_num, len(questions.list_of_questions))
for i in range(q_num):
    question = questions.list_of_questions[i]
    question.display_question()
    while True:
        user_answer = int(input("Ваш ответ (1-5): "))
        if 1 <= user_answer <= 5:
          break
        else:
          print("Пожалуйста, введите число от 1 до 5.")
    if question.check_answer(user_answer):
        print("Правильно!\n")
        sum+=1
    else:
        print(f"Неправильно. Правильный ответ: {question.ans}\n")
print(f'Correct answers count = {sum}\nThe percentage of correct answers is {sum/q_num*100}%')
```
